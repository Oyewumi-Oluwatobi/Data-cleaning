```python
import pandas as pd


# reading the database
data = pd.read_csv(r"C:\Users\user\Desktop\movies.csv")

# printing the top 10 rows
display(data.head ())
```

```python
data.drop_duplicates()
```

```python
data['YEAR'] = data ['YEAR'].str.replace('[^0-9]', '')
data['ONE-LINE'] = data ['ONE-LINE'].str.replace('[^a-zA-Z0-9]', ' ')
data['GENRE'] = data['GENRE'].str.strip('\n')
data
```

```python
data["YEAR"]=data["YEAR"].apply(lambda x : str(x))
data["YEAR"]=data["YEAR"].apply(lambda x : (x[0:4] + ", " + x[4:8]) if len(x)>4 else x)
data
```

```python
data[['directors', 'stars']] = data['STARS'].str.split('Stars:', expand = True)
data.drop('STARS', inplace = True,axis = 1)
data
```

```python
data.info ()
```

```python
data[['directors', 'stars']] = data[['directors', 'stars']].replace('[^a-zA-Z0-9]', '', regex = True)
data
```

```python
data['directors'] = data['directors'].str.replace('Director', '')
data
```

```python
data['VOTES'] = data['VOTES'].str.replace('[^0-9.]', '')
data
```

```python
data.reset_index(drop=True)
```

```python
data = data.drop_duplicates(subset = ['MOVIES','YEAR','GENRE', 'directors'])
data
```

```python
data.isnull().sum()
```

```python
data['Gross'].unique()
grossdata = data [data['Gross'].notnull()]
```

```python
grossdata.reset_index(drop=True)
```

```python
data = data.copy()

data['VOTES'] = pd.to_numeric(data['VOTES'], errors='coerce')

data['RATING'] = pd.to_numeric(data['RATING'], errors='coerce')
```

```python
data.info ()
```

```python
noratingdata = data[data['RATING'].isnull() | data['VOTES'].isnull()]
data = data.dropna(subset=['RATING', 'VOTES'])
data.drop('Gross', inplace=True, axis=1)
```

```python
grossdata = grossdata.copy()

grossdata['VOTES'] = pd.to_numeric(data['VOTES'], errors='coerce')

grossdata['RATING'] = pd.to_numeric(data['RATING'], errors='coerce')
```

```python
grossdata['VOTES']=grossdata['VOTES'].astype(str)
grossdata['Gross']=grossdata['Gross'].astype(str)

grossdata['VOTES']=grossdata['VOTES'].str.replace('[^0-9]', '')
grossdata['Gross']=grossdata['Gross'].str.replace('[^0-9]', '')
```

```python
grossdata = grossdata.copy()

grossdata['VOTES'] = pd.to_numeric(data['VOTES'], errors='coerce')

grossdata['RATING'] = pd.to_numeric(data['RATING'], errors='coerce')
```

```python
moviesdata = data
moviesdata.to_excel('moviesdata.xlsx', index=False)
noratingdata.to_excel('ratingdata.xlsx', index=False)
grossdata.to_excel('grossdata.xlsx', index=False)
moviesdata
```
